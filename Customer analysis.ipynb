```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
```

```python
df_trn = pd.read_csv("data/training.tsv", sep="\t", names=["id", "date", "email_status"])
df_trn.head()
```

```python
df_trn.dtypes
```

```python
df_trn["date"] = pd.to_datetime(df_trn["date"])
df_trn.head()
```

```python
df_trn.dtypes
```

```python
print(f"Total number of records in dataset: \n{df_trn.count()}")
```

```python
print(f"Total number of unique id's in dataset: {df_trn['id'].nunique()}")
```

```python
daily_orders_count = df_trn.groupby(["date"]).size().reset_index(name="counts")
daily_orders_count.head()
```
